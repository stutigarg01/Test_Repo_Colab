```python
# import the required libraries
# This library help us to handle HTTP request
import requests 
# import the pandas libraries to handle the tabular data
import pandas as pd
```

```python
# collect the data from API
# set the URL for json placeholder API to fetch user data
URL = 'https://jsonplaceholder.typicode.com/users'
# fetch the data from API
response = requests.get(URL)
```

```python
response
```

```python
# check if request was successful or not
if response.status_code ==200:
    data=response.json()
    # print the data
    print("fetch the json data")
    print(data)
else:
    print("Failed to the data.HTTP status code:",response.status_code)
    
```

```python
# transform the data into the tabular format 
# required data into the tables
# id, name, username, email, address, phone, website, company
[user['id'] for user in data]
```

```python
# fetch the id_data
Id_Data = [user['id'] for user in data]
```

```python
# fetch the id_data
Id_Data = [user['id'] for user in data]
# fetch the name_data
Name_Data = [user['name'] for user in data]
# fetch the username_data
Username_Data = [user['username'] for user in data]
# fetch the name_data
Email_Data = [user['email'] for user in data]
# fetch the name_data
Address_Data = [user['address'] for user in data]
# fetch the name_data
Phone_Data = [user['phone'] for user in data]
# fetch the name_data
Website_Data = [user['website'] for user in data]
# fetch the name_data
Company_Data = [user['company'] for user in data]
```
